Processing the data

```python
df['show_id'].value_counts().sum()
```

```python
df.set_index('show_id',inplace = True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df[['director','cast','country']] = df[['director','cast','country']].replace(np.nan, ' Unknown')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(subset = ['rating','duration','date_added'],inplace = True)
```

```python
df.isna().sum()
```

```python
df['date_added'] = df['date_added'].astype('datetime64[ms]')
```

```python
df.info()
```

```python
df['added_year'] = df['date_added'].dt.year
df['added_year']
```

```python
df['added_month'] = df['date_added'].dt.month_name()
df['added_month']
```
